```python
W = pd.DataFrame(w)
```

```python
learning_rate = 0.01
```

```python
Y = pd.DataFrame(y)
```

```python
test_num = 1000
```

```python
for i in range(test_num):
    cost = (X.dot(W) - Y).T
    W = W - (cost.dot(X) * learning_rate/len(X)).T
    print("test_num : ", i)
    print("slope : ",W.iloc[0])
    print("y-intercept : ",W.iloc[1])
```
